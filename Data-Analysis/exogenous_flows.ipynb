```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta
```

```python
exogenous_flows = np.load('../Data/all-exogenous.npy')
```

```python
exogenous_flows.shape
```

```python
start_date = datetime(2003, 1, 1)
end_date = datetime(2020, 8, 31)
date_range = [start_date + timedelta(days=i * 30) for i in range(len(exogenous_flows))]
```

```python
plt.figure(figsize=(17, 6))
plt.plot(date_range, exogenous_flows[:, 0], marker='o', linestyle='-', color='#34495E')
plt.title(f'Lake Superior Exogenous Flows', fontsize=15)

for year in range(start_date.year, end_date.year + 1):
    plt.axvline(datetime(year, 1, 1), color='red', linestyle='--', linewidth=0.8)

plt.ylabel('Net Flow (in meters)', fontsize=8.5)

plt.grid(True)
plt.show()
```
