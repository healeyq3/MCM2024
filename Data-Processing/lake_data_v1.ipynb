## Reading in data for each lake
note: if needed we can easily read all the river data the same way

```python
# Access the provided excel file with Lake & River Data
filePath = '../Data/Problem_D_Great_Lakes.xlsx'
file = pd.ExcelFile(filePath)

greatLakesData = {}

# Iterate through each sheet in the Excel file
for sheet_name in file.sheet_names:

    # Read each sheet into a DataFrame and store as a dictionary
    df = pd.read_excel(file, sheet_name)
    greatLakesData[sheet_name] = df
```

```python
### LAKE SUPERIOR ###

# Exctracting water level for Lake Superior 
superiorDepth = greatLakesData['Lake Superior'].iloc[5:, :]
superiorDepth.columns = superiorDepth.iloc[0].tolist()
superiorDepth = superiorDepth.iloc[1:, :]

# Adjsuting the way data frame is presented
## we want two columns: Time and Value

## melt the data frame and make 'Month' a categorical variable
s_melted = pd.melt(superiorDepth, id_vars=['Year'], var_name='Month', value_name='Value')
monthOrder = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
s_melted['Month'] = pd.Categorical(s_melted['Month'], categories=monthOrder, ordered=True)

## now sort the DataFrame by 'Year' and 'Month'
s_melted = s_melted.sort_values(by=['Year', 'Month'])

## combine Month and Year into one column
s_melted['Time'] = s_melted['Month'].astype(str) + '-' + s_melted['Year'].astype(str)

## create the final 1D DataFrame with two desired columns
superiorDepth = s_melted[['Time', 'Value']].reset_index(drop=True)
# superiorDepth
```

```python
def get_time(month_year: str) -> float:
    '''
    Conversion flow in m^3/s to m^3/month
    '''
    month = month_year[0:3]
    if month in ['Apr', 'Jun', 'Sep', 'Nov']:
        return 2.592 * 1e6
    elif month in ['Jan', 'Mar', 'May', 'Jul', 'Aug', 'Oct', 'Dec']:
        return 2.6784*1e6
    else:
        return 2.4192*1e6
```

```python
flow_time_vectors = []
for t in range(0, mary_flow.shape[0]):
    x_t = np.array([
        mary_flow['Value'][t] * get_time(mary_flow['Time'][t]),
        clair_flow['Value'][t] * get_time(clair_flow['Time'][t]),
        detroit_flow['Value'][t] * get_time(detroit_flow['Time'][t]),
        niagra_flow['Value'][t] * get_time(mary_flow['Time'][t]),
        lawrence_flow['Value'][t] * get_time(lawrence_flow['Time'][t])
    ])
    flow_time_vectors.append(x_t)
```

```python
np.save('../Data/flow-vectors.npy', flow_time_vectors)
```
